#### FUNCTIONS IN PYTHON :

#### CREATING A FUNCTION :


```python
# Example of function in python :
def is_even(num):
    print('Entered number is',num)
    '''
    This function returns if a given number is odd or even
    input = any valid 
    output = odd/even
    '''
    if type(num) == int:
        if num%2 == 0:
            return 'even'
        else:
            return 'odd'
    else:
        return 'Enter valid number !'
num = int(input('Enter the number :'))
result = is_even(num)
print(result)
```

###### 3.Keywords Arguments (named arguments) :

```python
#Ex.
def Student(name, age):
    print("Student Details :",name, age)

# default function call
Student('Jessa', 14)

# both keyword arguments
# both keyword arguments by changing their order
Student(age=21, name='Eren')

# 1 positional and 1 keyword
Student('Donald', age=13)
```

###### 4.Arbitrary arguments (variable-length arguments *args and **kwargs)

###### 1.Arbitrary positional arguments (*args)

```python
#Ex.
# function with variable-length arguments (*args)
def Percentage(*marks):
    sum = 0
    for i in marks:
        sum += i
    avg = sum / len(marks)
    print('*marks form a tuple :', marks)
    print("Average :", avg)

Percentage(56, 61, 73)
```

###### 2.arbitrary keyword arguments (**kwargs)

```python
# function with variable-length keyword arguments
def Percentage(**marks):
    for key,value in marks.items():
        print(key,'->',value) 
        
# pass multiple keyword arguments        
Percentage(English=56,Maths= 61,History=73)
```

```python
# declare global variable
message = 'Good Morning'

def greet():
    
     # declare local variable
    message = 'Good Night'
    print('Local :',message)

greet()

print('Global :',message)
```

##### Function returning another Function in Python

```python
# Python allows programmers to return a function from another function like it allows function as an argument.

def OuterFun():
    def ReturnFun():
        return 'This is return function !'
    print('OuterFun')
    return ReturnFun
rf  = OuterFun()
print(rf())
```

##### High order functions :

```python
# Ex.
# Here we have to create a function in that we pass list as input and function print the square of the input

def Sqaure(x):
    return x**2

def Cube(x):
    return x**3

# higer Order Function
def Transform(fun, li):
    output = []
    for i in li:
        output.append(fun(i))
    print(output)

li = [1,2,3,4]
Transform((lambda x:x**3),li) 
    
```

##### map() / filter() / reduce() functions:

```python
# square the items of a list 
x = map(lambda x:x**2,[9,8,7,6,5])
# map object convert into list (type converstion)
print(list(x)) 
```

```python
# odd/even labeling of list items
L = [3,4,5,6,67,78,8]
list(map(lambda x:'even' if x%2==0 else 'odd',L))
```

```python
# fetch names from a list of dict

users = [
    {
        'name':'Saitama',
        'age':25,
        'gender':'male'
    },
    {
        'name':'Eren',
        'age':23,
        'gender':'male'
    },
    {
        'name':'Mikasa',
        'age':22,
        'gender':'female'
    }
]

print(list(map(lambda users:users['name'],users)))
print(list(map(lambda users:users['age'],users)))
```

```python
# Add two lists using map and lambda
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]
 
result = map(lambda x, y: x + y, numbers1, numbers2)
print(list(result))
```
