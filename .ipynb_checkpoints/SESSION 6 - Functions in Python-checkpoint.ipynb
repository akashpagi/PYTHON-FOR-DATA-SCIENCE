#### FUNCTIONS IN PYTHON :

#### CREATING A FUNCTION :


```python
# Example of function in python :
def is_even(num):
    print('Entered number is',num)
    '''
    This function returns if a given number is odd or even
    input = any valid 
    output = odd/even
    '''
    if type(num) == int:
        if num%2 == 0:
            return 'even'
        else:
            return 'odd'
    else:
        return 'Enter valid number !'
num = int(input('Enter the number :'))
result = is_even(num)
print(result)
```

##### TYPES OF ARGUMENTS IN FUNCTION :

###### 1.Default Argument :

```python
#Ex.1
def Student(Name,age,Acadmic_Year=2023):
    print("Student Details :",Name,age,Acadmic_Year)

#not passing a school value (default value is used)    
Student('Eren',21)

# passign all arguments
Student('Nobita',15,2022)
```

###### 2.Positional Arguments :

```python
def Multi(a, b):
    return a*b
Multi(3,5)

def add(x,y=4):
    z = x + y
    print('Addition is', z)
add(10)
```

```python
#Note: ote: In the positional argument number and position of arguments must be matched. 
#If we change the order, then the result may change. Also, If we change the number of arguments, we will get an error.
def add(a, b):
    print(a + b)
add(105, 561, 4)
```

###### 3.Keywords Arguments (named arguments) :

```python
#Ex.
def Student(name, age):
    print("Student Details :",name, age)

# default function call
Student('Jessa', 14)

# both keyword arguments
# both keyword arguments by changing their order
Student(age=21, name='Eren')

# 1 positional and 1 keyword
Student('Donald', age=13)
```

###### 4.Arbitrary arguments (variable-length arguments *args and **kwargs)

###### 1.Arbitrary positional arguments (*args)

```python
#Ex.
# function with variable-length arguments (*args)
def Percentage(*marks):
    sum = 0
    for i in marks:
        sum += i
    avg = sum / len(marks)
    print('*marks form a tuple :', marks)
    print("Average :", avg)

Percentage(56, 61, 73)
```

###### 2.arbitrary keyword arguments (**kwargs)

```python
# function with variable-length keyword arguments
def Percentage(**marks):
    for key,value in marks.items():
        print(key,'->',value) 
        
# pass multiple keyword arguments        
Percentage(English=56,Maths= 61,History=73)
```
