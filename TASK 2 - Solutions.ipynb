```python
#Q2 Solution :

a = int(input("Enter the 1st angle :"))
b = int(input("Enter the 2nd angle :"))
c = int(input("Enter the 3rd angle :"))

if a + b + c == 180 and a!=0 and b!=0 and c!=0:
    print('The entered angles can form triangle')
else:
    print('The entered angles can not form triangle')
```

```python
#Q6 Solution :
# Factorial of the given number :
num  = int(input('Enter the number for factorial :'))

fac = 1

while(num > 0):
    fac = fac * num
    num  = num - 1
print('Factorial is',fac)
```

```python
#Q12. Solution :

num = int(input('Enter the number for checking the prime number or not :'))
if num<2:
    print(num ,'is not prime number')
else:   
    for i in range(2, num):
        if num%i == 0:
            print(num ,'is not prime number')
            break   
    else:
        print(num,"is prime number")
    
        
```
