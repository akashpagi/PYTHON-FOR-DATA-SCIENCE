#### OOP (Object-Oriented Programming) :
- Object-oriented programming (OOP) is a **programming paradigm based on the concept of "objects"**. 

- The object **contains both data and code***: Data in the form of properties (often known as attributes), and code, in the form of methods (actions object can perform).

- An object-oriented **paradigm is to design the program using classes and objects**. 
- Python programming language supports different programming approaches like functional programming, modular programming. 

- One of the popular approaches is object-oriented programming (OOP) to **solve a programming problem is by creating objects**



- Python OOP concepts
  - An object has the following two characteristics:
      - **Attribute**
      - **Behavior**

  - For example, A Car is an object, as it has the following properties:
    - name, price, color as attributes
    - breaking, acceleration as behavior
  - One important aspect of OOP in Python is to **create reusable code using the concept of inheritance**. This concept is also known as **DRY** (Don't Repeat Yourself).

#### What is a Class and Objects in Python?
#### Class :

#### Object :
- An **object is an instance of a class**.

- Object **represents the base class** from where all classes in python are derived.

- Objects have two characteristics: 
    - They have **states and behaviors** (object has attributes and methods attached to it) 
    - Attributes represent **its state, and methods represent its behavior**. Using its methods, we can modify its state.

```python
# Ex.

class Atm:
    #constructor
    def __init__(self):
        self.pin = ''
        self.balance = 0
        #print('constructor is executed automatically at the time of object creation.')
        self.menu()
        
    # adding functionality :
    def menu(self):
        user_input = input('''
        Hi How can help you ?
        1. Press 1 to create pin
        2. Press 2 to change pin
        3. Press 3 to check balance
        4. Press 4 to withdraw
        5. Press Anything to exist     
        ''')
        if user_input == '1':
            # create pin
            self.create_pin()
            
        elif user_input == '2':
            # change pin
            self.change_pin()
        
        elif user_input == '3':
            # check balance
            self.check_balance()
        
        elif user_input == '4':
            # withdraw
            self.withdraw()
        
        else:
            exit()
            
    def create_pin(self):
        user_pin = int(input('Enter your pin :'))
        self.pin = user_pin
        
        user_balance =  int(input('Enter your balance :'))
        self.balance = user_balance
        
        print('Pin created successfully !')
        self.menu()
        
    # change pin functionality of atm
    def change_pin(self):
        old_pin = int(input('Enter your existing pin for changing the pin :'))
        
        if old_pin == self.pin:
            # change the pin
            new_pin = int(input('Enter new pin :'))
            self.pin = new_pin
            print('Atm pin change successfully')
            self.menu()
        else:
            print('Invalid pin ! Plz try again !')
            self.menu()
            
            
    ## checking balance functionality of atm
    def check_balance(self):
        user_pin = int(input('Enter your pin :'))
        
        if user_pin == self.pin:
            print('Your bank balance is', self.balance)
            self.menu()
        else:
            print('Invalid pin ! Plz try again !')
            self.menu()
            
    ## Witdraw functionality of atm
    def withdraw(self):
        user_pin = int(input('Enter your pin :'))
        
        if user_pin == self.pin:
            # allow to withdraw 
            amount = int(input('Enter the amount :'))
            if amount <= self.balance:
                self.balance = self.balance - amount
                print('Withdrawl successful ! Your remaining Balance is',self.balance)
            else:
                print('Something is worng here !')
        else:
            print('Invalid pin ! Plz try again !')
            self.menu()
    
  
        
```

#### Self Keyword in Python :

```python
# Ex.
class Student:
    # Constructor
    def __init__(self, name , age):
        print("'self' address :",id(self))
        self.name = name
        self.age = age
    
    # self points to the current object
    def show(self):
        # access instance variable using self
        print(self.name ,self.age)

# Creating first object
Stu_obj = Student('Akash',21)
Stu_obj.show()
print("'Stu_obj' address :",id(Stu_obj))
print('Address of Object is equal to address of self in the method')
```

#### Fraction Class  , __str__ , __add__, __sub__ , __mul__, __truediv__ :

```python
# fraction has numrator and denometor

class fraction:
    # parameterized constructor
    def __init__(self,x,y):
        self.num = x
        self.den = y
        
    def __str__(self): #we you print the object
        return '{}/{}'.format(self.num,self.den)

    def __add__(self,other):
        new_num = self.num*other.den + other.num*self.den
        new_den = self.den * other.den
        return '{}/{}'.format(new_num,new_den)
    
    def __sub__(self,other):
        new_num = self.num*other.den - other.num*self.den
        new_den = self.den * other.den
        return '{}/{}'.format(new_num,new_den)
    
    def __mul__(self,other):
        new_num = self.num*other.num 
        new_den = self.den * other.den
        return '{}/{}'.format(new_num,new_den)

    def __truediv__(self,other):
        new_num = (self.num*other.den) 
        new_den = (self.den * other.num)
        return '{}/{}'.format(new_num,new_den)
```

```python
fr1= fraction(3,4)
print(fr1)
```

```python
fr2 = fraction(1,2)
print(fr2)
```

```python
# Addition of fraction fr1 and fr2
print(fr1 + fr2)
```

```python
# Subtraction of fraction fr1 and fr2
print(fr1 - fr2)
```

```python
# Multiplication of fraction fr1 and fr2
print(fr1 * fr2)
```

```python
# Division of fraction fr1 and fr2
print(fr1 / fr2)
```
